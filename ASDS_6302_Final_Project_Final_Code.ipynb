## Handling missing values

```python
# Drop columns that have more than 80% missing values
threshold = 0.8
drop_cols = missing_summary[missing_summary['Missing %'] > threshold * 100].index.tolist()
print("Dropping columns (too many missing):", drop_cols)
df = df.drop(columns=drop_cols)


for col in df.columns:
    if df[col].dtype == 'object':
        # Fill categorical with mode
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:
        # Fill numeric with mean
        df[col].fillna(df[col].mean(), inplace=True)
```

## Detect Outliers

```python
box_cols = [
    'Overall Score',
    'Student Achievement Score',
    'School Progress Score',
    '% Economically Disadvantaged',
    '% EB/EL Students'
]

# Convert to numeric
num_df = df[box_cols].apply(lambda col: pd.to_numeric(col, errors='coerce'))

# Keep only columns with at least 2 unique (non-NaN) values
valid_cols = [c for c in box_cols if num_df[c].dropna().nunique() >= 2]

print("Columns used for boxplot:", valid_cols)

# ---- Outlier detection (IQR) ----
for col in valid_cols:
    s = num_df[col].dropna()

    Q1 = s.quantile(0.25)
    Q3 = s.quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = s[(s < lower_bound) | (s > upper_bound)]
    print(f"{col}: {len(outliers)} outliers")

# ---- Single combined boxplot ----
plt.figure(figsize=(10, 6))
sns.boxplot(data=num_df[valid_cols])
plt.title("Boxplot of Selected Numeric Features (Outlier Detection)")
plt.xticks(rotation=45)
plt.ylabel("Value")
plt.tight_layout()
plt.show()
```

# Data Preprocessing before EDA

```python
drop_cols = [
    # Campus-level identifiers
    "Campus Number",
    "Campus",

    # School support / funding labels
    "Support Label",
    "Public Education Grant",

    # Political district fields (not useful for ML)
    "US Congress District",
    "US Congress Election District",
    "TX House District",
    "TX House Election District",
    "TX Senate District",
    "TX Senate Election District",

    # Misc campus enrollment type (not helpful)
    "Campus Enrollment Type"
]

# Drop only columns that exist
df = df.drop(columns=[c for c in drop_cols if c in df.columns])
```

```python
df["Overall Rating Letter"] = df["Overall Rating"].str.strip()
rating_order = ["A", "B", "C", "D", "F"]
```

## Encoding Variables

```python
# Target Variable
rating_map = {"A": 5, "B": 4, "C": 3, "D": 2, "F": 1}
df["Overall_Rating_Num"] = df["Overall Rating Letter"].map(rating_map)
```

```python
# Clean Distinction Columns  Convert Earned/Not Earned to 1/0
distinction_cols = [
    "Distinction ELA/Reading",
    "Distinction Mathematics",
    "Distinction Science",
    "Distinction Soc Studies",
    "Distinction Progress",
    "Distinction Closing the Gaps",
    "Distinction Postsecondary Readiness"
]

distinction_cols = [c for c in distinction_cols if c in df.columns]

for col in distinction_cols:
    df[col] = df[col].replace({"Earned":1, "Not Earned":0})
    df[col] = pd.to_numeric(df[col], errors="coerce").fillna(0).astype(int)
```

```python
protect_cols = (
    distinction_cols +
    ["School Type", "Grades Served", "Charter"]
)

# Convert numeric-like columns safely, but avoid protected ones
for col in df.columns:
    if df[col].dtype == "object" and col not in protect_cols:
        converted = pd.to_numeric(df[col], errors="coerce")
        if converted.notna().mean() > 0.90:   # majority numeric
            df[col] = converted
```

```python
def classify_grade_span(grade_str):
    if pd.isna(grade_str):
        return "Unknown"

    s = grade_str.upper().replace(" ", "")

    # High School
    if "09" in s or "10" in s or "11" in s or "12" in s:
        return "High School"

    # Middle School
    if "06" in s or "07" in s or "08" in s:
        return "Middle School"

    # Elementary
    if any(x in s for x in ["PK", "KG", "EE", "01", "02", "03", "04", "05"]):
        return "Elementary"

    # K12
    if "PK12" in s or "K12" in s or "EE12" in s:
        return "K12 Campus"

    return "Other"

df["Grade_Level"] = df["Grades Served"].apply(classify_grade_span)

# Drop original
#df = df.drop(columns=["Grades Served"])
```

```python
# Create Total Distinctions & Distinction Category
df["Total Distinctions"] = df[distinction_cols].sum(axis=1)

df["Distinction Category"] = pd.cut(
    df["Total Distinctions"],
    bins=[-1, 0, 3, 7],
    labels=["No Distinctions", "Some Distinctions", "High Distinction Schools"]
)
```

```python
score_cols = [c for c in df.columns if "Score" in c and "Distinction" not in c]

for col in score_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Clean and convert percentage and student count columns
percent_cols = ['% Economically Disadvantaged', '% EB/EL Students']
for col in percent_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.replace('%', '', regex=False)
        df[col] = pd.to_numeric(df[col], errors='coerce') / 100

if 'Number of Students' in df.columns:
    df['Number of Students'] = df['Number of Students'].astype(str).str.replace(',', '', regex=False)
    df['Number of Students'] = pd.to_numeric(df['Number of Students'], errors='coerce')

print("Cleaned score columns:", score_cols)
print("Cleaned percentage and student count columns dtypes:")
print(df[percent_cols + ['Number of Students']].dtypes)
```

```python
# District size
df["District_Size_Category"] = pd.cut(
    df["Number of Students"],
    bins=[0, 300, 1000, 5000, 20000],
    labels=["Small", "Medium", "Large", "Very Large"]
)
```

```python
# Economic disadvantage level
df["EcoDis_Category"] = pd.cut(
    df["% Economically Disadvantaged"],
    bins=[0, 0.3, 0.6, 1.0],
    labels=["Low", "Medium", "High"]
)
```

```python
# Charter type categorical
df["Charter"] = df["Charter"].astype("category")

# School type categorical
df["School Type"] = df["School Type"].astype("category")
```

# Feature Engineer


## 1. Performance Gap Index:
Create a feature that captures how far a schools Achievement and Progress scores deviate from its Closing the Gaps score.
* Schools with large gaps tend to have inconsistent performance  highly predictive of AF outcomes.

```python
# Performance Gap Index: inconsistency between domains
df["Performance_Gap"] = (
    df["Student Achievement Score"]
    + df["School Progress Score"]
    + df["Academic Growth Score"]
) / 3 - df["Closing the Gaps Score"]
```

* Positive values  Achievement/Progress are much higher than Closing Gaps (common in B/C schools)
* Negative values  Closing Gaps unusually high compared to Achievement (rare)

Higher gap  Worse consistency  Predicts lower rating

## 2. Achievement Efficiency Ratio

Capture how effectively a school converts its student population into academic performance, controlling for size + economic disadvantage. (Strong signal for A school)

Fomula: \text{Efficiency} = \frac{\text{Overall Score}}{\text{Number of Students}} \times (1 - \text{% Economically Disadvantaged})

* Larger districts often perform differently than small districts.
* Economic disadvantage is strongly negatively correlated with all score metrics in your heatmap.
* This ratio captures doing more with less.

```python
# Achievement Efficiency Ratio
df["Achievement_Efficiency"] = (
    df["Overall Score"] / (df["Number of Students"] + 1)
) * (1 - df["% Economically Disadvantaged"])
```

- Higher value  School performs much better than expected given student load + socio-economic pressure.
- Strong indicator for A-rated districts.

Since already have Total Distiction but rate is more predictive

## Function for confusion metric and ROC Curves

```python
def plot_conf_mat(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(title)
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.tight_layout()
    plt.show()


def plot_multiclass_roc(y_true, proba, title, classes=(0,1,2,3,4)):
    # binarize true labels
    y_bin = label_binarize(y_true, classes=classes)
    # Compute ROC for each class and macro-average
    fpr = dict(); tpr = dict(); roc_auc = dict()
    for i in range(len(classes)):
        fpr[i], tpr[i], _ = roc_curve(y_bin[:, i], proba[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    # macro-average
    all_fpr = np.unique(np.concatenate([fpr[i] for i in range(len(classes))]))
    mean_tpr = np.zeros_like(all_fpr)
    for i in range(len(classes)):
        mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
    mean_tpr /= len(classes)
    macro_auc = auc(all_fpr, mean_tpr)

    plt.figure(figsize=(6,5))
    plt.plot(all_fpr, mean_tpr, label=f"Macro-average ROC (AUC = {macro_auc:.3f})")
    for i in range(len(classes)):
        plt.plot(fpr[i], tpr[i], alpha=0.3, label=f"Class {classes[i]} (AUC={roc_auc[i]:.2f})")
    plt.plot([0,1], [0,1], "k--", alpha=0.4)
    plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
    plt.title(title)
    plt.legend(loc="lower right", fontsize=8)
    plt.tight_layout()
    plt.show()

    return macro_auc
```
